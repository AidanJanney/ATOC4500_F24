```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from scipy.integrate import odeint
```

```python
def lorenz(state, t, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]
```

### Time array

```python
t = np.linspace(0, 50, 10000)
```

### Initial conditions: [x, y, z]

```python
initial_state1 = [1.0, 1.0, 1.0]
initial_state2 = [0, 2.0, 20.0]
```

### Parameters for single attractor and double attractor

Alternate example set of parameters to get one or two attractor sets:


params_single = (10.0, 28.0, 8.0/3)

params_double = (10.0, 48.0, 4.0)

Original:

params_single = (10.0, 24.4, 8.0/3)

params_double = (10.0, 24.8, 8.0/3)

```python
params_single = (10.0, 24.4, 8.0/3)

params_double =(10.0, 24.4, 8.0/3)
```

### Integrate the Lorenz equations

```python
trajectory_single1 = odeint(lorenz, initial_state1, t, args=params_single)
trajectory_double1 = odeint(lorenz, initial_state1, t, args=params_double)
trajectory_double2 = odeint(lorenz, initial_state2, t, args=params_double)
```

### Plot

```python
fig = plt.figure(figsize=(12, 6))

# Single attractor 
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot(trajectory_single1[:, 0], trajectory_single1[:, 1], trajectory_single1[:, 2])
ax1.set_title('Single Attractor')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

# Double attractors
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot(trajectory_double1[:, 0], trajectory_double1[:, 1], trajectory_double1[:, 2], label='Attractor 1')
ax2.plot(trajectory_double2[:, 0], trajectory_double2[:, 1], trajectory_double2[:, 2], label='Attractor 2')
ax2.set_title('Double Attractors')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.legend()

plt.show()
```
